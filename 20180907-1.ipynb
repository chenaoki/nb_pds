```python
import numpy as np
import os, glob
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
src_path = "/mnt/recordings/SimulationResult/20170707-2/"
src_frame = "05400/"
```

```python
out_path = "/mnt/recordings/SimulationResult/20180907-1/"
out_frame = "cell_0000/"
if not os.path.exists(out_path): os.mkdir(out_path)
if not os.path.exists(out_path+out_frame): os.mkdir(out_path+out_frame)
```

## Check pacing wave form 

```python
files = sorted(glob.glob(src_path+"*/v.npy"))
v = np.array( [ np.load(f) for f in files ])[:,0,0]

plt.plot(v[-1000:-1])
```

## Broaden

```python
files = glob.glob(src_path+src_frame+"*.npy")
```

```python
X = np.zeros((200*200), dtype=np.float64)
for f in files:
    x = np.load(f)
    X[:] = x[0,0]
    np.save(f.replace(src_path+src_frame, out_path+out_frame), X)
```

```python
vmem = np.zeros((200,200), dtype=np.float64)
vmem[:,:] = v[-1]
np.save(out_path+"vmem_0000.npy", vmem)
```

```python
phie = np.zeros((200,200), dtype=np.float64)
np.save(out_path+"phie_0000.npy", phie)
```

```python
plt.subplot(121)
plt.imshow(vmem, vmax=vmem.max(), vmin=vmem.min())
plt.colorbar()

plt.subplot(122)
plt.imshow(phie, vmax=phie.max(), vmin=phie.min())
plt.colorbar()
```
